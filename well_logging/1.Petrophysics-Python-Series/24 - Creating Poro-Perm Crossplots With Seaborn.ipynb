# Creating Poro-Perm Crossplots with Seaborn

## Import the Required Libraries

```python
import seaborn as sns
import pandas as pd
```

### Read the CSV Data

```python
df = pd.read_csv("Data/15_9-19A-CORE.csv", na_values=' ')
```

```python
df.head()
```

### Create a Simple Scatter Plot

```python
sns.scatterplot(x=df['CPOR'], y=df['CKHL']);
```

### Change the Y-Axis to Logarithmic

```python
p = sns.scatterplot(x=df['CPOR'], y=df['CKHL']);
p.set(yscale='log');
```

### Add a Third Variable for Colour

```python
p = sns.scatterplot(x=df['CPOR'], y=df['CKHL'], hue=df['CGD'], palette='YlOrRd', s=100);
p.set(yscale='log');
```

### Tidy up the Plot Labels

```python
p = sns.scatterplot(x=df['CPOR'], y=df['CKHL'], hue=df['CGD'], palette='YlOrRd', s=100)
p.set(yscale='log')
p.set_ylabel('Core Permeability (mD)')
p.set_xlabel('Core Porosity (%)');
```

```python
p = sns.scatterplot(x=df['CPOR'], y=df['CKHL'], hue=df['CGD'], palette='YlOrRd', s=100)
p.set(yscale='log')
p.set_ylabel('Core Permeability (mD)', fontsize=12, fontweight='bold')
p.set_xlabel('Core Porosity (%)', fontsize=12, fontweight='bold');

```

### Change the Figure Size

```python
#Setup the figure size for the plot. This will also change the theme of the plot
sns.set(rc = {'figure.figsize':(15,8)})

p = sns.scatterplot(x=df['CPOR'], y=df['CKHL'], hue=df['CGD'], palette='YlOrRd', s=100)
p.set(yscale='log')
p.set_ylabel('Core Permeability (mD)', fontsize=12, fontweight='bold')
p.set_xlabel('Core Porosity (%)', fontsize=12, fontweight='bold');
```
